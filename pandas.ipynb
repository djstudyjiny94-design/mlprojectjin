```python
a={'name':['Tom','nick','krish','jack'],'age':[20,21,19,18]}
df=pd.DataFrame(a)
df
```

```python
df['name']=df['name'].str.upper()
df
```

```python
df
```

```python
new_row = {'name': 'SAM', 'age':None }
df = pd.concat([df, pd.DataFrame([new_row])])
df
```

```python
new_row={'name':'jiny','age':None}
df=pd.concat([df,pd.DataFrame([new_row])], ignore_index=True)
df
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce').fillna(0)
df
```

```python
df.fillna({'name':'jiny',
           'age':df['age'].mean()}, inplace=True)
df
```

```python
df.fillna({'name': 'jiny', 'age': 0}, inplace=True)
df
```

```python
df.rename(columns={'name':'name_id'},inplace=True)
```
